```python
# weather 
# we need two library 1. from bs4 import BeautifulSoup & 2.requests
# User-Agent = Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36


def weather(city="Noida"):
    url = f'https://www.google.com/search?q=weather+{city}'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    temperature = soup.select_one('span#wob_tm').text
    unit = soup.select_one('div.vk_bk.wob-unit span.wob_t').text
    description = soup.select_one('span#wob_dc').text
    return f"{temperature} {unit} {description}"
```

```python
# voice speech_ to_text
# we need two library 1.SpeechRecognition & 2.PyAudio

def speech_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        audio = recognizer.listen(source)
    try:
        voice_data = recognizer.recognize_google(audio)
        print(voice_data)
        return voice_data.lower()
    except sr.UnknownValueError:
        print("Error")
        return "Error"
    except sr.RequestError:
        print("RequestError")
        return "RequestError"
```

```python
# voice text_to_speech
# we need two library 1.pyttsx3

def text_to_speech(text):
    engine = pyttsx3.init()
    rate = engine.getProperty('rate')           # speaking rate
    engine.setProperty('rate', 'rate-50')
    engine.say(text)
    engine.runAndWait()
```
